Считаем .csv файлы, в которых хранится информация о станциях метро

```python
import csv


# Чтение CSV файла
with open('data/stations.csv', mode ='r', encoding='utf-8') as f:
    csv_reader = csv.reader(f)
    stations_list = []
    
    #присвоим первую часть датасета
    for row in csv_reader:
       stations_list.append(row) 
print(stations_list)
```

```python
import csv


# Чтение CSV файла
with open('data/exits.csv', mode ='r', encoding='utf-8') as f:
    csv_reader = csv.reader(f)
    exist_list = []
    
    #присвоим первую часть датасета
    for row in csv_reader:
       exist_list.append(row) 
print(exist_list)
```

Выведем основные аргументы из массива для работы

```python
list_of_arguments = []
for row in stations_list[0]:
    for argument in row.split(', '):
        list_of_arguments.append(argument)

#выведем лист наших аргументов, т.е. характеристик квартир
print(list_of_arguments)
```

```python
list_of_arguments_exist = []
for row in exist_list[0]:
    for argument in row.split(', '):
        list_of_arguments_exist.append(argument)

#выведем лист наших аргументов, т.е. характеристик квартир
print(list_of_arguments_exist)
```

Напишем модуль, который согласно названию нашей станции позволит оценить наличие той или иной услуги на станции метро

```python
#Справочная информация по запросу для каждой станции метро и МЦД по сферам услуг рядом со станцией и удобствами для маломобильных пассажиров
name_station = input()
id_index = list_of_arguments.index('id')
line_index = list_of_arguments.index('line')
exists_index = list_of_arguments.index('num_exits')
bank_index = list_of_arguments.index('BANK')
battery_index = list_of_arguments.index('BATTERY')
candy_index = list_of_arguments.index('CANDY')
carrier_index = list_of_arguments.index('CARRIER')
coffee_index = list_of_arguments.index('COFFEE')
elevator_index = list_of_arguments.index('ELEVATOR')
flowers_index = list_of_arguments.index('FLOWERS')
food_index = list_of_arguments.index('FOOD')
gift_index = list_of_arguments.index('GIFT_SHOP')
info_index = list_of_arguments.index('INFO')
invalid_index = list_of_arguments.index('INVALID')
optics_index = list_of_arguments.index('OPTICS')
parking_index = list_of_arguments.index('PARKING')
print_index = list_of_arguments.index('PRINT')
theatre_index = list_of_arguments.index('THEATRE')
toilet_index = list_of_arguments.index('TOILET')
vending_index = list_of_arguments.index('VENDING')
bus_index = list_of_arguments_exist.index('bus')
trolleybus_index = list_of_arguments_exist.index('trolleybus')
tram_index = list_of_arguments_exist.index('tram')
statid_index = list_of_arguments_exist.index('station_id')
idishnik = []
arguments_for_station = []
for row in stations_list[1:]:
    if row[1] == name_station:
        idishnik.append(row[id_index])
        print(f'Станция: {name_station}')
        print(f'Линия {row[line_index]}')
        if row[bank_index] == '1':
            print(f'Банки около станции: есть')
        else:
            print(f'Банки около станции: нет')
        if row[battery_index] == '1':
            print(f'Станции для аренды пауэрбанков около станции: есть')
        else:
            print(f'Станции для аренды пауэрбанков около станции: нет')
        if row[candy_index] == '1':
            print(f'Магазины сладостей около станции: есть')
        else:
            print(f'Магазины сладостей около станции: нет')
        if row[carrier_index] == '1':
            print(f'Эскалатор на станции: есть')
        else:
            print(f'Эскалатор на станции: нет')
        if row[coffee_index] == '1':
            print(f'Кофейни около станции: есть')
        else:
            print(f'Кофейни около станции: нет')
        if row[elevator_index] == '1':
            print(f'Лифты на станции: есть')
        else:
            print(f'Лифты на станции: нет')
        if row[flowers_index] == '1':
            print(f'Цветочные магазины около станции: есть')
        else:
            print(f'Цветочные магазины около станции: нет')
        if row[food_index] == '1':
            print(f'Заведения общепита около станции: есть')
        else:
            print(f'Заведения общепита около станции: нет')
        if row[gift_index] == '1':
            print(f'Магазины подарков около станции: есть')
        else:
            print(f'Магазины подарков около станции: нет')
        if row[info_index] == '1':
            print(f'Стойки информации на станции: есть')
        else:
            print(f'Стойки информации на станции: нет')
        if row[invalid_index] == '1':
            print(f'Пандусы на станции: есть')
        else:
            print(f'Пандусы на станции: нет')
        if row[optics_index] == '1':
            print(f'Магазины оптики около станции: есть')
        else:
            print(f'Магазины оптики около станции: нет')
        if row[parking_index] == '1':
            print(f'Паркинг около станции: есть')
        else:
            print(f'Паркинг около станции: нет')
        if row[print_index] == '1':
            print(f'Копировальные услуги около станции: есть')
        else:
            print(f'Копировальные услуги около станции: нет')
        if row[theatre_index] == '1':
            print(f'Театры около станции: есть')
        else:
            print(f'Театры около станции: нет')
        if row[toilet_index] == '1':
            print(f'Туалеты на станции: есть')
        else:
            print(f'Туалеты на станции: нет')
        if row[vending_index] == '1':
            print(f'Вендинговые автоматы на станции: есть')
        else:
            print(f'Вендинговые автоматы на станции: нет')
marshrutibus = []
marshrutitrolley = []
marshrutitram = []
for rows in exist_list[1:]: 
    for i in idishnik:
        if i == rows[statid_index]:
            for y in rows[bus_index].split(', '):
                if y == '':
                    continue
                if y not in marshrutibus:
                    marshrutibus.append(y)
            for y in rows[trolleybus_index].split(', '):
                if y == '':
                    continue
                if y not in marshrutitrolley:
                    marshrutitrolley.append(y)
            for y in rows[tram_index].split(', '):
                if y == '':
                    continue
                if y not in marshrutitram:
                    marshrutitram.append(y)


print(f'Маршруты автобусов около станции: {str(marshrutibus)}')
print(f'Маршруты троллейбусов около станции: {str(marshrutitrolley)}')
print(f'Маршруты трамваев около станции: {str(marshrutitram)}')
```

А теперь выгрузим только данные из массива 

```python
characteristic_station = []
for row in stations_list[1:]:
    characteristic_station.append(row)
print(characteristic_station)
```

А теперь рассчитаем коэфициент оснащенности для инвалидов для всех веток метро, который показывет часть станций, которое имеет оснащение.

```python
# Словарь для хранения количества станций и оснащённых станций по каждой линии
total_stations = {}
equipped_stations = {}

# Индексы нужных столбцов
line_index = list_of_arguments.index('line')
invalid_index = list_of_arguments.index('INVALID')

# Проходим по всем строкам данных
for row in characteristic_station:
    line = int(row[line_index])
    invalid = int(row[invalid_index])
    if line > 16:
        continue
    if line not in total_stations:
        total_stations[line] = 0
        equipped_stations[line] = 0
    total_stations[line] += 1
    if invalid == 1:
        equipped_stations[line] += 1

# Считаем коэффициенты
x = []
y = []
for line in sorted(total_stations.keys()):
    coeff = equipped_stations[line] / total_stations[line]
    x.append(line)
    y.append(coeff)

# Строим график
plt.figure(figsize=(12, 6))
plt.plot(x, y, marker='o', linestyle='-', color='purple')
plt.title('Коэффициент оснащённости станций для инвалидов по веткам метро')
plt.xlabel('Номер ветки')
plt.ylabel('Коэффициент оснащённости')
plt.xticks(x)
plt.ylim(-0.15, 1.15)
plt.show()
```

А теперь рассчитаем коэффициен оснащенности станций объектами питания по веткам метро

```python
# Словари для подсчёта
total_stations = {}
stations_with_food = {}

# Индексы нужных столбцов
line_index = list_of_arguments.index('line')
candy_index = list_of_arguments.index('CANDY')
coffee_index = list_of_arguments.index('COFFEE')
food_index = list_of_arguments.index('FOOD')
vending_index = list_of_arguments.index('VENDING')

# Проходим по всем строкам данных
for row in characteristic_station:
    line = int(row[line_index])
    if line > 16:
        continue
    if line not in total_stations:
        total_stations[line] = 0
        stations_with_food[line] = 0

    total_stations[line] += 1

    # Проверяем, есть ли хотя бы один признак еды (1 = есть)
    has_food = (
        int(row[candy_index]) == 1 or
        int(row[coffee_index]) == 1 or
        int(row[food_index]) == 1 or
        int(row[vending_index]) == 1
    )

    if has_food:
        stations_with_food[line] += 1

# Подготавливаем данные для графика
x = []
y = []
for line in sorted(total_stations.keys()):
    coeff = stations_with_food[line] / total_stations[line]
    x.append(line)
    y.append(coeff)

# Строим график
plt.figure(figsize=(12, 6))
plt.plot(x, y, marker='o', linestyle='-', color='green')
plt.title('Коэффициент оснащённости станций объектами питания по веткам метро')
plt.xlabel('Номер ветки')
plt.ylabel('Коэффициент оснащённости')
plt.xticks(x)
plt.ylim(-0.15, 1.15)
plt.show()
```
